```python
import pandas as pd
import os
import shutil 
import random
```

```python
File_path = "covid-chestxray-dataset-master/metadata.csv"
Images = "covid-chestxray-dataset-master/images"
```

```python
df = pd.read_csv(File_path)
```

```python
df.head()
```

```python
# Number of rows 

print("Number of rows in the file contain are : ", df.shape[0])
```

```python
# Number of columns

print("Number of columns in the file contain are : ", df.shape[1])
```

```python
Target_dir = "DataSet/Covid"

if not os.path.exists(Target_dir):
    os.mkdir(Target_dir)
    print("Covid Folder Created")
```

```python
count = 0

for (i,row) in df.iterrows():
    if row["finding"] == "COVID-19" and row['view'] == "PA":
        filename = row["filename"]
        img_path = os.path.join(Images, filename)
        img_copy_path = os.path.join(Target_dir, filename)
        shutil.copy2(img_path, img_copy_path)
        
        print("Moving the images ", count)
        count+=1
        
```

```python
print("The Number of COVID-19 images with front view are : ",count)
```

```python
# collect 201 images from kaggle dataset


Kaggle_File_path = "chest_xray/chest_xray/train/NORMAL"
Target_Normal_Dir = "DataSet/Normal"
```

```python
# list of all images present in the directory

img_name = os.listdir(Kaggle_File_path)

random.shuffle(img_name)
```

```python
for i in range(201):
    image_name = img_name[i]
    image_path = os.path.join(Kaggle_File_path, image_name)
    target_path = os.path.join(Target_Normal_Dir, image_name)
    shutil.copy2(image_path, target_path)
    print("Copying Images : ", i)
```
